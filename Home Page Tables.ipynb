```python
import pandas as pd
```

```python
weather_data = pd.read_csv("weather_data_update.csv")
weather_data.head()
```

```python
list(weather_data.columns)
```

```python
weather_tables_df = weather_data[['city_name', 'country', 'city_time', 
                                  'date_one_time_one_temp_min', 'date_one_time_one_temp_max',
                                  'date_two_time_one_temp_min', 'date_two_time_one_temp_max', 
                                 'date_three_time_one_temp_min', 'date_three_time_one_temp_max', 
                                  'date_four_time_one_temp_min', 'date_four_time_one_temp_max', 
                                 'date_five_time_one_temp_min', 'date_five_time_one_temp_max' ]]
                                  #'five_day_min_avg', 'five_day_max_avg', 'total_avg_all__five_days' 
weather_tables_df.head()
```

```python
weather_tables_df['five_day_min_avg'] = ""
weather_tables_df['five_day_max_avg'] = ""
weather_tables_df['total_avg_all__five_days'] = ""

weather_tables_df.head()
```

```python
weather_tables_df['five_day_min_avg'] = ((weather_tables_df['date_one_time_one_temp_min'] + weather_tables_df['date_two_time_one_temp_min'] + weather_tables_df['date_three_time_one_temp_min'] + weather_tables_df['date_four_time_one_temp_min'] + weather_tables_df['date_five_time_one_temp_min'])/5)
weather_tables_df.head()
```

```python
weather_tables_df['five_day_max_avg'] = ((weather_tables_df['date_one_time_one_temp_max'] + weather_tables_df['date_two_time_one_temp_max'] + weather_tables_df['date_three_time_one_temp_max'] + weather_tables_df['date_four_time_one_temp_max'] + weather_tables_df['date_five_time_one_temp_max'])/5)
weather_tables_df.head()
```

```python
weather_tables_df['total_avg_all__five_days'] = ((weather_tables_df['five_day_max_avg'] + weather_tables_df['five_day_min_avg'])/2)
weather_tables_df.head()
```

```python
web_tables_df = weather_tables_df[['city_name', 'country', 'city_time','five_day_min_avg', 'five_day_max_avg', 'total_avg_all__five_days' ]]
web_tables_df.head()
```
